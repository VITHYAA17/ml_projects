```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style('whitegrid')
```

```python
val=df['Age'].median()
df['Age']=df['Age'].fillna(val)
```

```python
df.isna().sum()
```

```python
mode_embarked=df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(mode_embarked)
```

```python
df.isna().sum()
```

```python
df['Has_Cabin']=df['Cabin'].notna().astype(int)
```

```python
df.head()
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

UNIVARIATE ANALYSIS

```python
fig,axes = plt.subplots(2,3,figsize=(18,12))
fig.suptitle('Univariate analysis of categorical features', fontsize=16)

sns.countplot(ax=axes[0,0],x='Survived',data=df).set_title('Survival Distribution')
sns.countplot(ax=axes[0,1],x='Pclass',data=df).set_title('Passenger Class Distribution')
sns.countplot(ax=axes[0,2],x='Sex',data=df).set_title('Sex classification')
sns.countplot(ax=axes[1,0],x='Embarked',data=df).set_title('Port of embarkation')
sns.countplot(ax=axes[1,1],x='SibSp',data=df).set_title('Sibilings/spouses Abroad')
sns.countplot(ax=axes[1,2],x='Parch',data=df).set_title('Parents /Children Abraod')
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
print("Analysing numerical features")

fig,axes = plt.subplots(1,2,figsize=(16,4))
fig.suptitle('Univariate analysis of numerical features',fontsize=16)

sns.histplot(ax=axes[0],data=df,x='Age',kde=True,bins=10).set_title('Age Distribution')
sns.histplot(ax=axes[1],data=df,x='Fare',kde=True,bins=40).set_title('Fare Distribution')
plt.show()
```

Bivariate Analysis

```python
g = sns.FacetGrid(df,col='Survived',height=6)
g.map(sns.histplot,'Age',bins=25, kde=True)
plt.suptitle('Age Distribution by survival Status',y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(y='Fare' , data=df)
plt.title('Box plot of ticket fare')
plt.ylabel('Fare')
plt.show()
```

Feature engineering

```python
#combining features
df['FamilySize']=df['SibSp']+df['Parch']+1

#create is alone
df['IsAlone']=0
df.loc[df['FamilySize']==1,'IsAlone']=1
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Multivariate Analysis: Pclass and Sex vs Survived', fontsize=16)

sns.countplot(x='Pclass', hue='Survived', data=df, ax=axes[0])
axes[0].set_title('Survival by Passenger Class')

sns.countplot(x='Sex', hue='Survived', data=df, ax=axes[1])
axes[1].set_title('Survival by Sex')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df, split=True)
plt.title('Age Distribution by Sex and Survival Status')
plt.show()
```
